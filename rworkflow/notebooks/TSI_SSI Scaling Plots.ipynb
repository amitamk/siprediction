```python
print(np.min(np.concatenate([T1,T2,T3,T4,T5])))
print(np.max(np.concatenate([T1,T2,T3,T4,T5])))
print('')
print(np.min(np.concatenate([T1_1,T2_1,T3_1,T4_1])))
print(np.max(np.concatenate([T1_1,T2_1,T3_1,T4_1])))
print('')
print(np.min(np.concatenate([T1_2,T2_2,T3_2,T4_2])))
print(np.max(np.concatenate([T1_2,T2_2,T3_2,T4_2])))
print('')
print(np.min(np.concatenate([T1_3,T2_3,T3_3,T4_3])))
print(np.max(np.concatenate([T1_3,T2_3,T3_3,T4_3])))
```

```python
figuresPath = 'C:\\Users\Ami\Dropbox\Tese\Monografia\Figuras\\'

fig = plt.figure(figsize=(12,6))

gs = gridspec.GridSpec(1, 4)#, width_ratios=[3, 1]) 

plt.suptitle("RNNs' Output Data Range",fontsize='xx-large')
ax0 = plt.subplot(gs[0])
ax0.set_title ("30.5nm",fontsize='x-large')
parts = ax0.violinplot(np.concatenate([T1_1,T2_1,T3_1,T4_1]), showmeans=True, showmedians=True,
        showextrema=True)
# ax0.set_xlabel("Output (30.5nm)",fontsize='x-large')
ax0.set_ylabel("$W/m^2/nm$",fontsize='x-large')
ax0.grid()
ax0.tick_params(axis='x', labelsize='large' ) 
ax0.tick_params(axis='y', labelsize='large' ) 

for pc in parts['bodies']:
    pc.set_facecolor('#8888FF')
    pc.set_edgecolor('black')
    pc.set_alpha(1)

ax1 = plt.subplot(gs[1])
ax1.set_title ("48.5nm",fontsize='x-large')
parts = ax1.violinplot(np.concatenate([T1_2,T2_2,T3_2,T4_2]), showmeans=True, showmedians=True,
        showextrema=True)
# ax1.set_xlabel("Output (48.5nm)",fontsize='x-large')
ax1.set_ylabel("$W/m^2/nm$",fontsize='x-large')
ax1.grid()
ax1.tick_params(axis='x', labelsize='large' ) 
ax1.tick_params(axis='y', labelsize='large' ) 

for pc in parts['bodies']:
    pc.set_facecolor('#88FF88')
    pc.set_edgecolor('black')
    pc.set_alpha(1)

ax2 = plt.subplot(gs[2])
ax2.set_title ("121.5nm",fontsize='x-large')
parts = ax2.violinplot(np.concatenate([T1_3,T2_3,T3_3,T4_3]), showmeans=True, showmedians=True,
        showextrema=True)
# ax2.set_xlabel("Input Feature",fontsize='x-large')
ax2.set_ylabel("$W/m^2/nm$",fontsize='x-large')
ax2.grid()
ax2.tick_params(axis='x', labelsize='large' ) 
ax2.tick_params(axis='y', labelsize='large' ) 

for pc in parts['bodies']:
    pc.set_facecolor('#FF8888')
    pc.set_edgecolor('black')
    pc.set_alpha(1)

ax3 = plt.subplot(gs[3])
ax3.set_title ("TSI",fontsize='x-large')
parts = ax3.violinplot(np.concatenate([T1,T2,T3,T4,T5]), showmeans=True, showmedians=True,
        showextrema=True)
# ax2.set_xlabel("Input Feature",fontsize='x-large')
ax3.set_ylabel("$W/m^2$",fontsize='x-large')
ax3.grid()
ax3.tick_params(axis='x', labelsize='large' ) 
ax3.tick_params(axis='y', labelsize='large' ) 

for pc in parts['bodies']:
    pc.set_facecolor('#888888')
    pc.set_edgecolor('black')
    pc.set_alpha(1)

plt.tight_layout()

plt.savefig(figuresPath+'Output_Range.pdf',dpi=200,format='pdf',padinch=0.1)
```
