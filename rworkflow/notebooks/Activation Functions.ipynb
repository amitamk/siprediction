```python
import math
import matplotlib.pyplot as plt
import numpy as np
outputPath = 'C:\\Users\Ami\Dropbox\Tese\Monografia\Figuras\\'
#import libs

import pandas as pd
import statsmodels.api as sm
```

```python
def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a

def tanh(x):
    a = []
    for item in x:
        a.append(2/(1+math.exp(-2*(item)))-1)
    return a    

def linear(x):
    a = []
    for item in x:
        a.append(item)
    return a    

def reLU(x):
    a = []
    for item in x:
        if item<0:
            a.append(0)
        else:
            a.append(item)
    return a    
```

```python
y = x
def sigmoidU(x):
    return 1/(1+math.exp(-x))

def sigmoidD(y):
    a = []
    for item in y:
        a.append(sigmoidU(np.array([item],float))*(1-sigmoidU(np.array([item],float))))
    return a

def tanhD(x):
    a = []
    for item in x:
        a.append(1)
    return a    

def linearD(x):
    a = []
    for item in x:
        a.append(1)
    return a    

def reLUD(x):
    a = []
    for item in x:
        if item<0:
            a.append(0)
        else:
            a.append(1)
    return a    
```
