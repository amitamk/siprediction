```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import dates
from datetime import *
```

```python
T = np.asarray(np.loadtxt('./partial_output_files/412D/T_412dias_6hours.csv'))
T1 = np.asarray(np.loadtxt('./partial_output_files/412D/T_30.5_412dias1days.csv'))
T2 = np.asarray(np.loadtxt('./partial_output_files/412D/T_48.5_412dias1days.csv'))
T3 = np.asarray(np.loadtxt('./partial_output_files/412D/T_121.5_412dias1days.csv'))
Time = np.asarray(np.loadtxt('./partial_output_files/412D/Time_412dias_6hours.csv'))
SN = np.asarray(np.loadtxt('SN.txt', delimiter=";"))
```

```python
print(date.fromordinal(int(Time[0])))
print(date.fromordinal(int(Time[-1])))
T.shape[0]
```

```python
dataT = []
for i in Time:
    dataT.append(str(date.fromordinal(int(i))))

datesT = dates.date2num(dataT)

hfmt = dates.DateFormatter('%Y/%m/%d')
```

```python
f, (ax1,ax2,ax3,ax4) = plt.subplots(4,1,figsize=(12,7))

ax = plt.gca()
ax.xaxis.set_major_formatter(hfmt)

ax1.set_title('Spectral and Total Solar Irradiance for 2011/09/14 to 2012/10/30', fontsize='xx-large')
ax1.plot(T1,'r',label="30.5nm")
ax1.set_ylabel(r'$W/m^2/nm$',fontsize='x-large')
ax1.set_xticklabels([])
ax1.tick_params(axis='y', labelsize='large')
ax1.legend(fontsize='large')

ax2.plot(T2,'g',label="48.5nm")
ax2.set_ylabel(r'$W/m^2/nm$',fontsize='x-large')
ax2.set_xticklabels([])
ax2.tick_params(axis='y', labelsize='large')
ax2.legend(fontsize='large')

ax3.plot(T3,'b',label="121.5nm")
ax3.set_ylabel(r'$W/m^2/nm$',fontsize='x-large')
ax3.set_xticklabels([])
ax3.tick_params(axis='y', labelsize='large')
ax3.legend(fontsize='large')

ax4.plot(datesT,T,'y',label="TSI")
ax4.set_ylabel(r'$W/m^2$',fontsize='x-large')
ax4.set_xlabel('Time',fontsize='x-large')
# ax4.set_xticks(np.arange(0,1651,30))
# ax4.set_xticklabels([])
ax4.tick_params(axis='y', labelsize='large')
ax4.tick_params(axis='x', labelsize='large')
ax4.legend(fontsize='large')

plt.savefig('./Figures/Spectral and Total Irradiance.pdf',dpi=200,format='pdf', pad_inches=0.1)
```
